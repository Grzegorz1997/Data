# Column headers do not match

```python
dete_survey = pd.read_csv("dete_survey.csv", na_values = "Not Stated")
```

```python
dete_survey_updated = dete_survey.drop(dete_survey.columns[28:49],axis =1)
```

```python
tafe_survey_updated = tafe_survey.drop(tafe_survey.columns[17:66], axis =1)
```

# Empty values in dete_survey has been changed to string "Not Stated" to keep the same format
# From both dataframes we droped columns which are not needed for the analysis

```python
dete_survey_updated.columns = dete_survey_updated.columns.str.lower().str.strip().str.replace(" ", "_")
```

```python
tafe_survey_updated.rename(columns = {
    'Record ID': 'id',
    'CESSATION YEAR': 'cease_date',
    'Reason for ceasing employment': 'separationtype',
    'Gender. What is your Gender?': 'gender',
    'CurrentAge. Current Age': 'age',
    'Employment Type. Employment Type': 'employment_status',
    'Classification. Classification': 'position',
    'LengthofServiceOverall. Overall Length of Service at Institute (in years)': 'institute_service',
    'LengthofServiceCurrent. Length of Service at current workplace (in years)': 'role_service'}, inplace = True)
```

# Columns needed for analysis were transformed to same format

```python
dataframes = [dete_survey_updated,tafe_survey_updated]
```

```python
for df in dataframes:
    df["separationtype"] = df["separationtype"].astype(str).fillna("Not Stated")
    print(df["separationtype"].value_counts(dropna= False))
```

```python
dete_resignations = dete_survey_updated[dete_survey_updated["separationtype"].str.lower().str.contains("resignation")].copy()
```

```python
tafe_resignations = tafe_survey_updated[tafe_survey_updated["separationtype"].str.lower().str.contains("resignation")].copy()
```

# Replaced NaN in tafe_survey_updated for "Not Stated"
# filtered dataframes by resignation separation type

```python
dete_resignations["cease_date"].value_counts()
```

```python
dete_resignations["cease_date"] = dete_resignations["cease_date"].astype(str)
```

```python
dete_resignations["cease_date"] = dete_resignations["cease_date"].str.extract(r"([0-9]{4})", expand = False).astype(float)
```

```python
dete_resignations["institute_service"] = dete_resignations["cease_date"] - dete_resignations["dete_start_date"]
```

# Added new column in dete_resignations to show number of years at the company

```python
dete_dissatisfaction_list = [
    'job_dissatisfaction',
    'dissatisfaction_with_the_department',
    'physical_work_environment',
    'lack_of_recognition',
    'lack_of_job_security',
    'work_location',
    'employment_conditions',
    'work_life_balance',
    'workload']
tafe_dissatisfaction_list = [
    'Contributing Factors. Dissatisfaction',
    'Contributing Factors. Job Dissatisfaction']
```

```python
for col in dete_dissatisfaction_list:
    print(dete_resignations[col].value_counts(dropna=False))
```

```python
for col in tafe_dissatisfaction_list:
    print(tafe_resignations[col].value_counts(dropna = False))
```

```python
def update_vals(val):
    if pd.isnull(val):
        return np.nan
    elif val == '-':
        return False
    else:
        return True
```

```python
def check_dissatisfied(row):
    if any(row):
        return True
    elif all(pd.isnull(row)):
        return np.nan
    else:
        return False
```

```python
tafe_resignations[tafe_dissatisfaction_list] = tafe_resignations[tafe_dissatisfaction_list].applymap(update_vals)
tafe_resignations["dissatisfied"] = tafe_resignations[tafe_dissatisfaction_list].apply(check_dissatisfied,axis =1)
tafe_resignations_up = tafe_resignations.copy()
```

```python
dete_resignations["dissatisfied"] = dete_resignations[dete_dissatisfaction_list].apply(check_dissatisfied, axis =1)
dete_resignations_up = dete_resignations.copy()
```

# changed analysed reason columns in tafe_resignations to bool

```python
dete_resignations_up["institute"] = "DETE"
tafe_resignations_up["institute"] = "TAFE"
```

```python
combined = pd.concat([dete_resignations_up,tafe_resignations_up])
```

```python
combined_updated = combined.dropna(thresh = 500, axis= 1)
```

# Added column to identical data source in combined dataframe
# combined dataframes and dropped column with more than 500 null values

```python
combined_updated["institute_service"].value_counts()
```

```python
def clean_service_val(val):
    if pd.isnull(val):
        return np.nan
    elif val < 3:
        return "New"
    elif (val >= 3) & (val <= 6):
        return "Experienced"
    elif (val >= 7) & (val <= 10):
        return "Established"
    else:
        return "Veteran"
```

```python
combined_updated["service_cat"] = combined_updated["institute_service"].astype(str).str.extract(r"(\d+)").astype(float)
```

```python
combined_updated["service_cat"] = combined_updated["service_cat"].apply(clean_service_val)
```

```python
combined_updated["service_cat"]
```

```python
combined_updated['dissatisfied'] = combined_updated['dissatisfied'].fillna(False)
```
